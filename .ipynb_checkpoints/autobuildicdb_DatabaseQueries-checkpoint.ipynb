```python
import pandas as pd
import numpy as np
from pandas import ExcelFile
from pandas import ExcelWriter
import urllib.request
import requests, zipfile, io
from datetime import datetime
from xlrd import xldate
import pyodbc
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, delete, insert, select, func, sql
from sqlalchemy.types import Integer
import urllib
import pprint

#Change Pandas display settings to show all columns
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', None)
#pd.set_option('display.max_rows', 500)
```

```python
# Configure and connect to SQL Server database

pp = pprint.PrettyPrinter(indent=10)

params = 'DRIVER={ODBC Driver 13 for SQL Server};' \
         'SERVER=5SQFPQ2\SQLEXPRESS;' \
         'PORT=1433;' \
         'DATABASE=Test40;' \
         'Trusted_Connection=yes;'
            
params = urllib.parse.quote_plus(params)

engine = create_engine('mssql+pyodbc:///?odbc_connect=%s' % params)

metadata = MetaData()
  
#Create Table objects
wells = Table('WELLS', metadata, autoload=True, autoload_with=engine)
wellsuserfieldsvalues = Table('t_WellsUserFieldsValues', metadata, autoload=True, autoload_with=engine)
wellsuserfields = Table('t_WellsUserFields', metadata, autoload=True, autoload_with=engine)
datalithostrat = Table('DATA_Lithostrat', metadata, autoload=True, autoload_with=engine)
projects = Table('PROJECTS', metadata, autoload=True, autoload_with=engine)
wellqueries = Table('WELLQUERIES', metadata, autoload=True, autoload_with=engine)
projectwells = Table('PROJECTWELLS', metadata, autoload=True, autoload_with=engine)

connection = engine.connect()
                       
#pp.pprint(repr(wells))
```

<b>Join WELLS, t_WellsUserFields and t_WellsUserFieldsValues</b>

```python
sql = '''
SELECT ufv.f_WellId, w.name, ufv.f_FieldID, uf.f_FieldName, ufv.f_StringValue
FROM dbo.t_WellsUserFieldsValues AS ufv
JOIN dbo.WELLS AS w
ON w.well_id = ufv.f_WellId
JOIN dbo.t_WellsUserFields AS uf
ON uf.f_FieldId = ufv.f_FieldID
'''
pd.read_sql_query(sql, engine)
```

# DATA_GeochemPyrolysi001

```python
#See which IC Data Types exist in f_tableid/dest_table no. 531: DATA_GeochemPyrolysi001
#Join tablenames and intervalcolumns

sql = '''

SELECT 
    ta.tabletype, ta.tablename, ta.tabledescription, ta.f_tableid, 
    ty.dest_table, ty.title, ty.data_type
FROM dbo.tablenames AS ta
JOIN dbo.INTERVALCOLUMNS AS ty
    ON ta.f_tableid = ty.dest_table
WHERE ta.tablename = 'DATA_CO2'
ORDER BY ta.tablename, ty.title;

'''

pd.read_sql_query(sql, engine)
```

```python
# See which columns use this attr value in DATA_Lithostrat
# attr = '{"ZoneColour":-1,"ZoneColourIsIpAuto":true,"EventSymbolId":0,"IsLocked":false,"OriginalZoneIndex":0}'

sql = ''' 

SELECT DISTINCT(DATA_Lithostrat.data_type), INTERVALCOLUMNS.dest_table, INTERVALCOLUMNS.description
FROM DATA_Lithostrat
JOIN INTERVALCOLUMNS
ON DATA_Lithostrat.data_type = INTERVALCOLUMNS.data_type
WHERE attr = '{"ZoneColour":-1,"ZoneColourIsIpAuto":true,"EventSymbolId":0,"IsLocked":false,"OriginalZoneIndex":0}'

'''

#DATA_Lithostrat.data_type, DATA_Lithostrat.attr



pd.read_sql_query(sql, engine)
```

```python
#Join wells, geochempyrolysis and intervalcolumns

sql = '''
SELECT w.name AS "Well Name", 
    ic.data_type AS "IC Data Type",
    ic.description AS "Description", 
    gp1.top_depth AS "Depth", 
    gp1.legend AS "Value"
FROM dbo.WELLS AS w
INNER JOIN dbo.DATA_GeochemPyrolysi001 AS gp1
ON w.well_id = gp1.well_id
INNER JOIN dbo.INTERVALCOLUMNS AS ic
ON gp1.data_type = ic.data_type
WHERE dest_table = 531

'''

#dest_table 531 for DATA_GeochemPyrolysi001

pd.read_sql_query(sql, engine).head()
```

```python
# DATA_GeochemPyrolysi001 will show all data_types vertically.
# Need to convert rows to columns, one per data_type.
# Can do this in pandas after the import using the opposite of pd.melt
# e.g. df_explo_nondefaultattributes = pd.melt(df_explo_nondefaultattributes, id_vars='well_id')
# See https://stackoverflow.com/questions/22127569/opposite-of-melt-in-python-pandas

# How to do the same in SQL? A self join?

sql = '''
SELECT
    gp1.top_depth AS "Depth", 
    gp1.legend AS "Value",
    gp1.data_type
FROM dbo.DATA_GeochemPyrolysi001 AS gp1
INNER JOIN dbo.DATA_GeochemPyrolysi001 AS gp2
ON gp1.pk_index = gp2.pk_index
AND gp1.data_type = gp2.data_type


'''

pd.read_sql_query(sql, engine).head()
```
