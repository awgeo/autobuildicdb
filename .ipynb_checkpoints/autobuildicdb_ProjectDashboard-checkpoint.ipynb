```python
# Return dbo.WELLS table
# See https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_sql.html#pandas.read_sql

sql = '''
SELECT *
FROM dbo.WELLS
'''

df_icwells = pd.read_sql_query(sql, engine)
df_icwells.head()
```

```python
# Count the number of wells in each main_area/location

print('Exploration well count in each main_area:')
explo_area_well_count = Counter(df_icwells['location'])
print(explo_area_well_count, '\n')

# print('Development well count in each main_area:')
# dev_area_well_count = Counter(df_dev['main_area'])
# print(dev_area_well_count, '\n')

# print('Total well count in each main_area:')
# total_area_well_count = explo_area_well_count + dev_area_well_count
# print(total_area_well_count)
```

```python
main_areas = []
explo_wells_per_area = []
#all_wells_per_area = []

for item in explo_area_well_count.most_common(3):
    main_areas.append(item[0].capitalize())
    explo_wells_per_area.append(item[1])

# for item in total_area_well_count.most_common(3):
#     all_wells_per_area.append(item[1])
    
print(main_areas)
print(explo_wells_per_area)
# print(all_wells_per_area)
```

```python
# Pie plots

labels = main_areas
explode = [0.1, 0.15, 0]

fig1, ax1 = plt.subplots()
#fig2, ax2 = plt.subplots()

ax1.pie(explo_wells_per_area, labels=labels, explode=explode, shadow=True, colors=pal,
       startangle=90, autopct='%1.1f%%',
       wedgeprops={'edgecolor': 'black'})

# ax2.pie(all_wells_per_area, labels=labels, explode=explode, shadow=True, colors=pal,
#        startangle=90, autopct='%1.1f%%',
#        wedgeprops={'edgecolor': 'black'})

ax1.set_title('Exploration wells per NCS region', fontsize=20)
ax1.legend(loc='lower right', prop={'size': 12})
#ax2.set_title('Exploration and Development wells per NCS region', fontsize=20)

#plt.savefig('Plots/pie_explo_and_dev_wells_per_region.png')
plt.show()
```

```python
# Join WELLS, t_WellsUserFields and t_WellsUserFieldsValues

sql = '''
SELECT ufv.f_WellId, w.name, ufv.f_FieldID, uf.f_FieldName, ufv.f_StringValue
FROM dbo.t_WellsUserFieldsValues AS ufv
JOIN dbo.WELLS AS w
ON w.well_id = ufv.f_WellId
JOIN dbo.t_WellsUserFields AS uf
ON uf.f_FieldId = ufv.f_FieldID
--- WHERE (ufv.f_WellId = 99) AND (ufv.f_FieldID = 37)
ORDER BY w.name
'''
temp_df = pd.read_sql_query(sql, engine)
temp_df.head(56)
```
