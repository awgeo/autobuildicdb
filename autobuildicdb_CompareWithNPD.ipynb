```python
# Assign dbo.wells to df_ic

sql = ''' SELECT *FROM dbo.WELLS '''

df_ic = pd.read_sql_query(sql, engine)
# df_ic.drop(columns='pk_index', inplace=True)

print(df_ic.shape)
df_ic.iloc[[0, -1]]
```

```python
ic_dict = dict(zip(df_ic.well_id, df_ic.name))
ic_dict
```

```python
# Difference between two dictionaries. Create two sets.
set1 = set(npd_dict.items())
set2 = set(ic_dict.items())

# Symmetric difference operation of sets
# https://docs.python.org/3/library/stdtypes.html#frozenset.symmetric_difference
wells_not_in_ic = set1 ^ set2
len(wells_not_in_ic)
```

```python
# Return df showing NPD wells not in IC
test = pd.DataFrame.from_dict(wells_not_in_ic)
test.columns = ['well_id', 'name']
test
```

```python
well_ids = test['well_id'].to_list()
filt = df_npd['NPDID wellbore'].isin(well_ids)
df_npd[filt]

# Then append this to the database
# 
```
